```python
from pprint import pprint

list_example = df['normalisasi']
list_example_term=[str(document).lower().split() for document in df['normalisasi']]
print('tokenized term:')
pprint(list_example_term)
```

```python
example_n=len(list_example)
print('n:')
print(example_n)
```

```python
from collections import Counter
from pprint import pprint

example_n=len(list_example)
print('n:')
print(example_n)
print()
list_example_term_t=[str(document).lower().split() for document in list_example]
print('tokenized term (t):')
pprint(list_example_term_t)
print()
list_example_term_tc =[list(Counter(document).items()) for document in list_example_term_t]
print('term count (tc):')
pprint(list_example_term_tc)
print()
list_example_term_tt =[sum([term[1]  for term in document]) for document in list_example_term_tc]
print('total term (tt):')
pprint(list_example_term_tt)
print()
counter_example_term_df =Counter([term[0] for document in list_example_term_tc for term in document])
print('document frequency (df):')
pprint(list(counter_example_term_df.items()))
print()
```

```python
import numpy as np

def calculate_example_tfidf(indx,t,tc,tt,n,df):
  tf=tc/tt
  idf=np.log10((n)/(df))
  tfidf=tf*idf
  return [indx,t,tc,tt,tf,df,idf,tfidf]

for indx,document in enumerate(list_example_term_tc):
  for t,tc in document:
    print(calculate_example_tfidf(indx
                                  ,t
                                  ,tc
                                  ,list_example_term_tt[indx]
                                  ,example_n
                                  ,counter_example_term_df[t]))
```

```python
#IDF calculation using log n
import numpy as np

def calculate_example_tfidf(di,t,tc,tt,n,df):
  tf=tc/tt
  idf=np.log((n)/(df))
  tfidf=tf*idf
  return [di,t,tc,tt,tf,n,df,idf,tfidf]

for di,document in enumerate(list_example_term_tc):
  for t,tc in document:
    print(calculate_example_tfidf(di
                                  ,t
                                  ,tc
                                  ,list_example_term_tt[di]
                                  ,example_n
                                  ,counter_example_term_df[t]))
```

```python
# IDF Calculation using sklearn approach

import numpy as np

def calculate_example_tfidf_adjusted(di,t,tc,tt,n,df):
  list_result=[]
  tf=tc/tt
  idf=np.log((1+n)/(1+df)) +1
  tfidf=tf*idf
  return [di,t,tc,tt,tf,n,df,idf,tfidf]

for di,document in enumerate(list_example_term_tc):
  for t,tc in document:
    print(calculate_example_tfidf_adjusted(di
                                           ,t
                                           ,tc
                                           ,list_example_term_tt[di]
                                           ,example_n
                                           ,counter_example_term_df[t]))
```

```python
# convert TF data into a dataframe for more view options
import pandas as pd

list_example_tfidf=[calculate_example_tfidf_adjusted(
    indx
    ,t
    ,tc
    ,list_example_term_tt[indx]
    ,example_n
    ,counter_example_term_df[t])
for indx,document in enumerate(list_example_term_tc) for t,tc in document]


df_example_tf=pd.DataFrame(list_example_tfidf
                           ,columns=['di','t','tc','tt','tf','n','df','idf','tfidf'])

df_example_tf
```
